#### Read counts and match with ploidy data file

This script counts the reads of every fastq file and creates an output. 

It additionally matches names and ploidy with  **CHDO_pop.csv** to make sure no descripencies

```python
import sys

#sys.path.append('/home/faske/g/anaconda3/envs/py34/lib/python3.4/site-packages')
sys.path.append('/data/gpfs/assoc/parchmanlab/tfaske/anaconda3/envs/py36/lib/python3.6/site-packages')
sys.path.append("/data/gpfs/assoc/parchmanlab/tfaske/ipynb/include_utils")

import ipyparallel as ipp
import os, time
from os import environ
import include_utils as u
import pandas as pd
import numpy as np
import scipy as sp
import random
```

```python
fq_dir = '/data/gpfs/assoc/denovo/tfaske/sagebrush/fastq'
```

```python
cd $fq_dir
```

```python
fq_files = []
files = !find . -type f -name '*.F.fq.gz'
files = [os.path.abspath(x) for x in files]
for x in files:
    fq_files.append(x)
fq_files = sorted(fq_files)
len(fq_files),fq_files[0]
```

#### Count  reads for all fq

```python
fq_names = []
fq_reads = []
for f in fq_files:
    environ['f'] = f
    r = !zcat $f | awk '{s++}END{print s/4}'
    name = f.split('.F')[0]
    name = name.split('/')[8]
    #print(name)
    #print(int(r.n))
    fq_names.append(name)
    fq_reads.append(int(r.n))
```

```python
len(fq_reads), fq_reads[:5]
```

```python
len(fq_names), fq_names[0]
```

```python
read_dict = {"fq":fq_names,'reads':fq_reads}
read_df = pd.DataFrame(read_dict)
read_df.head()
#len(cov_df)
```

```python
data_dir = '/data/gpfs/assoc/denovo/tfaske/sagebrush/data'
```

```python
read_out = os.path.join(data_dir,'fq_read.csv')
read_df.to_csv(path_or_buf=read_out)
```

```python
%%R
require(tidyverse)
library(ggforce)
library(ggsci)
setwd('/data/gpfs/assoc/denovo/tfaske/sagebrush/data')

ARTR_pop <- read.csv('ARTR_pop.csv')
read_df <- read.csv('fq_read.csv')
read_df$fq <- as.character(read_df$fq)

## rename AV2_CF
read_df$fq[which(as.character(read_df$fq) == 'AV2_CF_1')] <- 'AV4_CF_1'
read_df$fq[which(as.character(read_df$fq) == 'AV2_CF_2')] <- 'AV4_CF_2'
read_df$fq[which(as.character(read_df$fq) == 'AV2_CF_3')] <- 'AV4_CF_3'

print(ARTR_pop[1:5,])
print(read_df[1:5,])
```

```python
%%R 
#### make files match ####
ssp <- c('tridentata','vaseyana','wyomingensis')
ARTR_pop <- ARTR_pop[which(ARTR_pop$subspecies %in% ssp),]

#rename ssp 
ARTR_pop$ssp <- NA
ARTR_pop$ssp[which(ARTR_pop$subspecies == 'tridentata')] <- 'AT'
ARTR_pop$ssp[which(ARTR_pop$subspecies == 'vaseyana')] <- 'AV'
ARTR_pop$ssp[which(ARTR_pop$subspecies == 'wyomingensis')] <- 'AW'

#make ssp_pop for ARTR_pop 
ARTR_pop$ssp_pop <- paste0(ARTR_pop$ssp,as.character(ARTR_pop$Ploidy),'_',ARTR_pop$our.abbrev)
 
#make ssp_pop for read_df 
read_df$ssp <- sapply(read_df$fq, function(s) unlist(strsplit(as.character(s),'_'))[1])
read_df$pop <- sapply(read_df$fq, function(s) unlist(strsplit(as.character(s),'_'))[2]) 
read_df$ID <- sapply(read_df$fq, function(s) unlist(strsplit(as.character(s),'_'))[3]) 
read_df$ssp_pop <- paste(read_df$ssp,read_df$pop,sep='_')        
                      
#ARTR_reads <- left_join(read_df,ARTR_pop)
```

```python
%%R
ARTR_pop[1:5,]
read_df[1:5,]
print(length(unique(read_df$ssp_pop)))
print(length(unique(ARTR_pop$ssp_pop)))

print(length(which(unique(read_df$ssp_pop) %in% unique(ARTR_pop$ssp_pop))))
print(ARTR_pop$ssp_pop[-which(unique(ARTR_pop$ssp_pop) %in% unique(read_df$ssp_pop))])
print(unique(read_df$ssp_pop)[-which(unique(read_df$ssp_pop) %in% unique(ARTR_pop$ssp_pop))])


### ALLL MATCH ###
```

```python
%%R

ARTR_reads <- left_join(read_df,ARTR_pop,by='ssp_pop')
ARTR_reads[1:5,]
names(ARTR_reads)

ARTR_reads <- ARTR_reads[c(17,12,5,6,2,7,4,8,3,14,15)]
names(ARTR_reads) <- c('ssp','ploidy','pop','ID','All','ssp_pop','ssp_ploid',
                       'Name','read_count','Latitude','Longitude')
write.csv(ARTR_reads,'ARTR_reads.csv',row.names=F)
```

### summarize and plot read counts 

```python
%%R 

ARTR_reads <- read.csv('ARTR_reads.csv')

ARTR_reads_sum <- ARTR_reads %>%
  group_by(ssp_ploid) %>%
  summarise(mean = mean(read_count, na.rm = TRUE), 
            sd = sd(read_count, na.rm = TRUE), 
            n = n()) %>%
  mutate(se = sd / sqrt(n),
         ci = 1.96 * se)

ARTR_reads_sum
```
